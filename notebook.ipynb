- Bagaimana pola penggunaan sepeda berdasarkan musim?
- Apakah ada perbedaan jumlah penggunaan sepeda antara hari kerja dan akhir pekan?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Load dataset day.csv
day_data = pd.read_csv('data/day.csv')

# Load dataset hour.csv
hour_data = pd.read_csv('data/hour.csv')

# Tampilkan beberapa baris pertama untuk memastikan data telah dimuat dengan benar
day_data.head()
hour_data.head()
```

### Assessing Data

```python
# Cek missing values
print("Missing values di day_data:")
print(day_data.isnull().sum())

print("Missing values di hour_data:")
print(hour_data.isnull().sum())

# Deskripsi statistik
print("Deskripsi statistik untuk day_data:")
print(day_data.describe())

print("Deskripsi statistik untuk hour_data:")
print(hour_data.describe())
```

### Cleaning Data

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Visualisasi distribusi jumlah penggunaan sepeda berdasarkan musim
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='cnt', data=day_data)
plt.title('Jumlah Penggunaan Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Penggunaan Sepeda')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Winter', 'Spring', 'Summer', 'Fall'])  # Ganti angka dengan nama musim
plt.show()
```

```python
# Visualisasi perbedaan jumlah penggunaan sepeda antara hari kerja dan akhir pekan
plt.figure(figsize=(10,6))
sns.boxplot(x='workingday', y='cnt', data=day_data)
plt.title('Jumlah Penggunaan Sepeda: Hari Kerja vs Akhir Pekan')
plt.xlabel('Hari Kerja (1) vs Akhir Pekan (0)')
plt.ylabel('Jumlah Penggunaan Sepeda')
plt.xticks(ticks=[0, 1], labels=['Akhir Pekan', 'Hari Kerja'])  # Ganti angka dengan label yang sesuai
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:Bagaimana Pola Penggunaan Sepeda Berdasarkan Musim?

```python
# Kode untuk visualisasi pola penggunaan sepeda berdasarkan musim
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='cnt', data=day_data)
plt.title('Jumlah Penggunaan Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Penggunaan Sepeda')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

### Pertanyaan 2:Apakah Ada Perbedaan Jumlah Penggunaan Sepeda Antara Hari Kerja dan Akhir Pekan

```python
# Kode untuk visualisasi perbandingan hari kerja dan akhir pekan
plt.figure(figsize=(10,6))
sns.boxplot(x='workingday', y='cnt', data=day_data)
plt.title('Jumlah Penggunaan Sepeda: Hari Kerja vs Akhir Pekan')
plt.xlabel('Hari Kerja (1) vs Akhir Pekan (0)')
plt.ylabel('Jumlah Penggunaan Sepeda')
plt.xticks(ticks=[0, 1], labels=['Akhir Pekan', 'Hari Kerja'])
plt.show()
```

## Conclusion

- Conclusion Pertanyaan 1:
  Berdasarkan analisis visualisasi, terlihat bahwa penggunaan sepeda cenderung paling tinggi pada musim panas (Summer), diikuti oleh musim semi (Spring) dan musim gugur (Fall). Penggunaan sepeda paling rendah terjadi pada musim dingin (Winter). Hal ini kemungkinan besar disebabkan oleh faktor cuaca, di mana musim panas lebih mendukung untuk bersepeda di luar ruangan dibandingkan musim dingin.
- Conclusion Pertanyaan 2:
  Berdasarkan analisis visualisasi, terlihat bahwa penggunaan sepeda cenderung lebih tinggi pada akhir pekan dibandingkan dengan hari kerja. Hal ini menunjukkan bahwa sepeda lebih sering digunakan untuk rekreasi pada akhir pekan, sementara pada hari kerja, penggunaan sepeda mungkin lebih difokuskan pada keperluan transportasi untuk bekerja atau ke sekolah.
